```python
zipfile.ZipFile('../input/whats-cooking/train.json.zip').extractall()
train_data = pd.read_json('/kaggle/working/train.json')
train_data.head(10)
```

```python
zipfile.ZipFile('../input/whats-cooking/test.json.zip').extractall()
test_data = pd.read_json('/kaggle/working/test.json')
```

<h2>Data Preparation:</h2>

```python
train_data["cuisine"].value_counts()
```

```python
train_data.isnull().sum()
```

```python
train_data["ingredients"] = train_data["ingredients"].apply(lambda x : ' '.join(map(str, x))).apply(str.lower)
test_data["ingredients"] = test_data["ingredients"].apply(lambda x : ' '.join(map(str, x))).apply(str.lower)
```

```python
train_data
```

```python
X_train = train_data["ingredients"]
y_train = train_data["cuisine"]
X_test = test_data["ingredients"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
```

```python
X_train = pd.DataFrame(data=X_train.toarray().astype(float),columns = vectorizer.get_feature_names_out())
X_train
```

```python
X_train.shape
```

```python
# Get the column names that are not numeric
non_numeric_cols = [col for col in X_train.columns if not col.isnumeric()]

# Drop the columns that are not in the non_numeric_cols list
X_train = X_train.drop(columns=[col for col in X_train.columns if col not in non_numeric_cols])
```

```python
X_train_columns = X_train.columns
```

```python
X_train
```

```python
X_train.shape
```

```python
vectorizer = CountVectorizer()
X_test = vectorizer.fit_transform(X_test)
```

```python
X_test = pd.DataFrame(data=X_test.toarray().astype(float),columns = vectorizer.get_feature_names_out())
```

```python
# Get the column names that are not numeric
non_numeric_cols = [col for col in X_test.columns if not col.isnumeric()]

# Drop the columns that are not in the non_numeric_cols list
X_test = X_test.drop(columns=[col for col in X_test.columns if col not in non_numeric_cols])
```

```python
X_test_columns = X_test.columns
```

<h2>Merge the columns of the train and test data so both have same features</h2>

```python
X_test_tmp = pd.DataFrame(np.zeros((X_train.shape[0], X_test.shape[1])), columns=X_test_columns)
X_train_tmp = pd.DataFrame(np.zeros((X_test.shape[0], X_train.shape[1])), columns=X_train_columns)
```

```python
# If a column is common in both we sum the values 
df1 = pd.concat([X_train, X_test_tmp], axis=1)
X_train = df1.groupby(df1.columns, axis=1).sum()
df2 = pd.concat([X_test, X_train_tmp], axis=1)
X_test = df2.groupby(df2.columns, axis=1).sum()
```

<h2>Split the Data to train and val so we can evaluate the model:</h2>

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, random_state = 42, test_size = 0.30)
```

<h2>Modeling</h2>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression(penalty='l2')
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_train)

print(accuracy_score(y_train, lr_pred))
```

```python
lr_pred_val = lr.predict(X_val)
print(accuracy_score(y_val, lr_pred_val))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, lr_pred_val))
```
